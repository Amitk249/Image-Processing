```python
# Step 2: Import libraries
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import requests

# Step 3: Load the BLIP model and processor
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")
```

```python
from google.colab import files
uploaded = files.upload()

# Step 4: Load the uploaded image
image_path = list(uploaded.keys())[0]
image = Image.open(image_path)
```

```python
text = "girl playing basketball "

# Step 6: Preprocess the image and prompt for the model
inputs = processor(image, text, return_tensors="pt")

# Step 7: Generate the caption
out = model.generate(**inputs)

# Step 8: Decode and print the generated caption
caption = processor.decode(out[0], skip_special_tokens=True)
print(caption)
```
